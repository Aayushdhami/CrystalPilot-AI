### 1. Install Dependencies and Load YOLO Model
We will use the Ultralytics implementation of YOLO for efficient real-time detection.

```python
!pip install ultralytics opencv-python-headless

import cv2
import numpy as np
from ultralytics import YOLO

# Load a lightweight YOLOv8 model
model = YOLO('yolov8n.pt')
```

### 2. Define Light Zones and IoT Simulation
We map specific coordinates in the camera view to Light IDs.

```python
# Simulated IoT Light Control
def send_iot_command(light_id, status):
    """Simulates sending a signal to an IoT controller (e.g., MQTT or REST API)"""
    print(f"[IOT SIGNAL] Light {light_id} is now {status.upper()}")

# Define zones (x_min, y_min, x_max, y_max) and corresponding light IDs
# We will use relative coordinates (0.0 to 1.0) to make it work on any camera resolution
LIGHT_ZONES = {
    "Light_A_Desk": [0.0, 0.0, 0.5, 1.0],    # Left half
    "Light_B_Sofa": [0.5, 0.0, 1.0, 1.0]     # Right half
}

active_lights = set()
```

### 2.1 Connect to Physical Hardware
Replace the code below with your board's IP address and API endpoint. This replaces the virtual print statements with real network signals.

```python
import requests

# Replace with the actual IP of your IoT lighting board
BOARD_IP = "192.168.1.100"

def send_iot_command(light_id, status):
    """Sends a real network command to the lighting board"""
    url = f"http://{BOARD_IP}/control"
    payload = {"light": light_id, "state": status}

    try:
        # Sending the signal to the physical sensor
        response = requests.post(url, json=payload, timeout=2)
        if response.status_code == 200:
            print(f"[SUCCESS] Physical {light_id} is now {status.upper()}")
        else:
            print(f"[FAILED] Board returned error: {response.status_code}")
    except Exception as e:
        print(f"[CONNECTION ERROR] Could not reach board at {BOARD_IP}: {e}")

print("Physical IoT Bridge Ready. Update BOARD_IP to your sensor's address.")
```

### 3. Motion Detection and Logic Loop
This function processes frames, detects people, and triggers the lights.

```python
def process_motion(frame):
    global active_lights
    h, w, _ = frame.shape
    results = model(frame, verbose=False)
    detected_in_zones = set()

    for r in results:
        for box in r.boxes:
            if int(box.cls[0]) == 0: # Person
                x1, y1, x2, y2 = box.xyxy[0].tolist()
                cx, cy = (x1 + x2) / 2 / w, (y1 + y2) / 2 / h # Normalized center

                for light_id, coords in LIGHT_ZONES.items():
                    if coords[0] <= cx <= coords[2] and coords[1] <= cy <= coords[3]:
                        detected_in_zones.add(light_id)

    for light_id in detected_in_zones:
        if light_id not in active_lights:
            send_iot_command(light_id, "on")
            active_lights.add(light_id)

    to_remove = [l for l in active_lights if l not in detected_in_zones]
    for lid in to_remove:
        send_iot_command(lid, "off")
        active_lights.remove(lid)

print("System Logic Updated to use Normalized Coordinates.")
```

### 6. Live Webcam Detection
Run the cell below to start your camera. It will detect people and print the IoT light commands live.

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

def take_photo(filename='photo.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture Frame';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)
  return filename

try:
  print("Starting Webcam for single frame analysis...")
  filename = take_photo()
  print('Saved to {}'.format(filename))

  # Process the captured image
  frame = cv2.imread(filename)
  process_motion(frame)

  # Show result (Fixed coordinate scaling)
  h, w, _ = frame.shape
  for light_id, coords in LIGHT_ZONES.items():
      x1, y1 = int(coords[0] * w), int(coords[1] * h)
      x2, y2 = int(coords[2] * w), int(coords[3] * h)
      cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)

  from google.colab.patches import cv2_imshow
  cv2_imshow(frame)

except Exception as err:
  print(f"Error: {err}")
```

### 7. Real-Time Video Stream & Detection
This cell will start a live stream. It detects people and triggers the light IoT signals while you are watching the video.

```python
import PIL.Image
import io
import time
from IPython.display import display, Javascript, clear_output
from google.colab.output import eval_js
from base64 import b64decode

def run_live_detection(save_to_file=True):
    # JavaScript for a persistent stream with an explicit stop mechanism
    js = Javascript('''
    async function runStream() {
        const div = document.createElement('div');
        const video = document.createElement('video');
        const btn = document.createElement('button');
        btn.textContent = 'Stop Detection';
        video.style.display = 'block';
        const stream = await navigator.mediaDevices.getUserMedia({video: true});

        document.body.appendChild(div);
        div.appendChild(video);
        div.appendChild(btn);
        video.srcObject = stream;
        await video.play();

        const canvas = document.createElement('canvas');
        canvas.width = video.videoWidth;
        canvas.height = video.videoHeight;

        window.streaming = true;
        btn.onclick = () => { window.streaming = false; stream.getVideoTracks()[0].stop(); div.remove(); };

        window.captureFrame = () => {
            canvas.getContext('2d').drawImage(video, 0, 0);
            return canvas.toDataURL('image/jpeg', 0.8);
        };

        return {width: video.videoWidth, height: video.videoHeight};
    }
    window.streamDims = runStream();
    ''')
    display(js)

    print("Live Detection Running. Video will be saved to 'detection_output.avi'.")

    out = None
    try:
        dims = eval_js('window.streamDims')
        if save_to_file:
            # Define codec and create VideoWriter object
            fourcc = cv2.VideoWriter_fourcc(*'XVID')
            out = cv2.VideoWriter('detection_output.avi', fourcc, 10.0, (dims['width'], dims['height']))

        while eval_js('window.streaming'):
            # Capture frame
            data = eval_js('window.captureFrame()')
            binary = b64decode(data.split(',')[1])

            # Convert to OpenCV
            img = PIL.Image.open(io.BytesIO(binary))
            frame = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)

            # Run detection logic
            process_motion(frame)

            # Draw zones for the recorded file visualization
            for light_id, coords in LIGHT_ZONES.items():
                h, w, _ = frame.shape
                x1, y1 = int(coords[0] * w), int(coords[1] * h)
                x2, y2 = int(coords[2] * w), int(coords[3] * h)
                cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 255, 0), 2)
                cv2.putText(frame, light_id, (x1 + 5, y1 + 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 0), 1)

            if out:
                out.write(frame)

            # Small sleep to stabilize the frame capture rate
            time.sleep(0.1)

    except Exception as e:
        print(f"Stream ended or interrupted.")
    finally:
        if out:
            out.release()
            print("Successfully saved video to 'detection_output.avi'")

run_live_detection()
```
